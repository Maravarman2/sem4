# MORPHOLGY TRANSFORMATION AND IMAGE SEGMENTATION

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

# DILATION & EROTION

```python
img=cv.imread("C:/Users/SN139/Downloads/shapes2.png")
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
_,mask=cv.threshold(img,220,225,cv.THRESH_BINARY)
kernal=np.ones((3,3),np.uint8)
dilation=cv.dilate(mask,kernal,iterations=1)
erode=cv.erode(mask,kernal,iterations=1)
titles=['image','mask','dilate','erode']
images=[img,mask,dilation,erode]
for i in range(len(images)):
    plt.subplot(2,2,i+1),plt.imshow(images[i])
    plt.title(titles[i])
plt.imshow
```

# Morphological opening-erosion followed by dilation

```python
open=cv.morphologyEx(img,cv.MORPH_OPEN,kernal)
```

```python
cv.imshow("img",open)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Morphological closing-erosion followed by dilation

```python
cv.imshow("ORIGINAL",img)
cv.imshow('MORPH CLOSE',open)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Morphological Gradien-erosion,dilation,difference of both 

```python
open=cv.morphologyEx(img,cv.MORPH_GRADIENT,kernal)
```

# Display-Morhological Gradient

```python
cv.imshow("ORIGINAL",img)
cv.imshow('MORPH GEADIENT',open)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Top Hat-Opening and Different of Original and Opening

```python
tophat=cv.morphologyEx(img,cv.MORPH_TOPHAT,kernal)
```

```python
cv.imshow("ORIGINAL",img)
cv.imshow('MORPH_TOPHAT',tophat)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Black Hat Opp of Top Hat

```python
blackhat=cv.morphologyEx(img,cv.MORPH_BLACKHAT,kernal)
```

```python
cv.imshow("ORIGINAL",img)
cv.imshow('MORPH_BLACKHAT',blackhat)
cv.waitKey(0)
cv.destroyAllWindows()
```
